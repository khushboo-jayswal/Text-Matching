```python
std = std.drop([2,151])
std.isna().sum()
```

```python
cus = data['List of Causes']
cus
```

```python
cus.isna().sum()
```

```python
np.where(cus.isna())
```

```python
cus = cus[:255]
cus.isna().sum()
```

For Installation of Package: RUN Command --> %pip install fuzzywuzzy

```python
from fuzzywuzzy import fuzz
```

```python
Match = []
for i in std:
  for j in cus:
    Match.append(fuzz.partial_token_sort_ratio(i,j))
```

```python
result = np.array(Match).reshape(len(std),len(cus))
```

```python
fresult=pd.DataFrame(data = result ,columns = cus, index = std)
fresult
```

```python
fresult.to_csv('Final_Result.csv')
```
